```python
#import libraries

import pandas as pd
import numpy as np
```

```python
#read in csv dataset and create dataframe with it

salaries_df = pd.read_csv('Levels_Fyi_Salary_Data.csv')
```

```python
#see preview of dataset
salaries_df.head()
```

```python
#view all columns in dataset

salaries_df.columns
```

```python
#begin splitting dataset by interesting cuts

company_df = salaries_df[["timestamp", 
                          "company",
                          "title",
                          "totalyearlycompensation", 
                          "location",
                          "cityid",
                          "basesalary",
                          "stockgrantvalue",
                          "bonus",
                          "yearsatcompany"]]

gender_df = salaries_df[["timestamp", 
                         "gender",
                         "company",
                         "title",
                         "totalyearlycompensation",
                         "location",
                         "cityid",
                         "basesalary",
                         "stockgrantvalue",
                         "bonus",
                         "yearsatcompany"]]

education_df = salaries_df[["timestamp",
                            "company",
                            "title",
                            "totalyearlycompensation",
                            "location",
                            "cityid",
                            "basesalary",
                            "stockgrantvalue",
                            "bonus",
                            "yearsatcompany",
                            "Masters_Degree",
                            "Bachelors_Degree",
                            "Doctorate_Degree",
                            "Highschool",
                            "Some_College"]]
```

## Cleaning Company DF

```python
#begin cleaning individual datasets starting with null salaries

company_df
null_base_salaries = company_df[company_df['basesalary'] == 0]
null_base_salaries
```

```python
#confirming there are no null values for total yearly comp field

null_total_comp = company_df[company_df['totalyearlycompensation'] == 0]
null_total_comp
```

```python
#viewing null values for company field, ran again after dropping n/a to confirm there are no nulls now for that field

null_company = pd.isnull(company_df["company"])
null_company

company_df[null_company]
```

```python
#drop company null values

company_df = company_df.dropna(subset=['company'])
company_df
```

```python
#reset index
company_df.reset_index(drop=True)
```

```python
#export company df to own excel file for tableau ingest
company_df.to_excel("company_dataframe.xlsx")
```

## Cleaning Gender DF

```python
gender_df
```

```python
#drop n/a on gender
gender_df = gender_df.dropna(subset=["gender"])
```

```python
#reset index
gender_df.reset_index(drop=True)
```

```python
#export gender df to own excel file for tableau ingest
gender_df.to_excel("gender_dataframe.xlsx")
```

## Cleaning Education DF

```python
education_df
```

```python
#identify rows with at least 1 education listed
ed_listed = (education_df['Masters_Degree'] == 1) | (education_df['Bachelors_Degree'] == 1) | (education_df['Doctorate_Degree'] == 1) | (education_df['Highschool'] == 1) | (education_df['Some_College'] == 1)
```

```python
# Filter the DataFrame based on null values
new_education_df = education_df[ed_listed]
```

```python
# Display the result. Appears that 1/2 the dataset doesn't have education listed
new_education_df
```

```python
#reset index

new_education_df.reset_index(drop=True)
```

```python
#export to excel file
new_education_df.to_excel("education_df.xlsx")
```
